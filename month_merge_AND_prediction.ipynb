```python
import numpy as np
import pandas as pd
```

```python
#load file
df1 = pd.read_csv('rating_m.csv')
# df1 = df1.set_index(pd.to_datetime(df1['date'],format='%Y%m%d'))
# df1 = df1.drop(['date'], axis=1)
df1['date'] = pd.to_datetime(df1['date'],format='%Y%m%d')
df1 = df1.drop(['Unnamed: 0'], axis=1)
# df = df.dropna()
df1.head()
```

```python
df2 = pd.read_csv('DATA_Company.csv')
# df2 = df2.set_index(pd.to_datetime(df2['date']))
# df = df.drop(['date'], axis=1)
df2['date'] = pd.to_datetime(df2['date'])
df2 = df2.drop(['Unnamed: 0'], axis=1)
df2.head()
```

```python
universe2 = list(set(df2.equity.values))
universe1 = list(set(df1.Ticker.values))
```

```python
print(len(universe2) , len(universe1))
```

```python
df_comp = pd.DataFrame()
for ticker in universe2:
    df_ = df2[df2.equity == ticker]
    df_['CUR_MKT_CAP'] = df_['CUR_MKT_CAP'].fillna(method='ffill')
    df_comp = pd.concat([df_comp, df_])
    
```

```python
df2 = df_comp.dropna()
```

```python
df2.head()
```

```python
df = pd.merge(df1,df2, left_on =('date','Ticker'), right_on = ('date','equity'))
```

```python
df = df.drop(['equity'], axis=1)
df = df.drop(['rate'], axis=1)
df = df.rename(columns={'r':'rating'})
df.head()
```

```python
df['zscore'] = 1.2*df['WORKING_CAPITAL']/df['BS_TOT_ASSET'] + 1.4*df['BS_PURE_RETAINED_EARNINGS']/df['BS_TOT_ASSET'] + 3.3*df['EBIT']/df['BS_TOT_ASSET'] + 0.6*df['CUR_MKT_CAP']/df['BS_TOT_ASSET'] + 0.999*df['IS_SALES_AND_SERVICES_REVENUES']/df['BS_TOT_ASSET']
```

```python
df = df.drop(['BS_TOT_ASSET', 'EBIT','BS_PURE_RETAINED_EARNINGS', 'CUR_MKT_CAP',
       'IS_SALES_AND_SERVICES_REVENUES', 'RSK_BB_IMPLIED_CDS_SPREAD',
       'RSK_BB_ISSUER_DEFAULT', 'BB_1YR_DEFAULT_PROB', 'PX_LAST', 'rating', 'WORKING_CAPITAL'], axis=1)
```

```python
#universe with length > 8
len_comp = []
universe = []
for ticker in universe2:
    df_ = df[df.Ticker == ticker]
    len_comp.append(len(df_))
    if len(df_) > 11:
        universe.append(ticker)
```

```python
# AR example
from statsmodels.tsa.ar_model import AR

df_pred = pd.DataFrame()

for ticker in universe:
    df1 = df[df.Ticker == ticker]
    
    data1 = df1['BS_TOT_LIAB2'].iloc[0:-4]
    data2 = df1['BS_TOT_LIAB2'].iloc[0:-3]
    data3 = df1['BS_TOT_LIAB2'].iloc[0:-2]
    data4 = df1['BS_TOT_LIAB2'].iloc[0:-1]
    pres_lia = []
    for data in (data1, data2, data3, data4):
        model = AR(data)
        model_fit = model.fit()
        lia_prediction = model_fit.predict(len(data), len(data))
        pred_lia = pd.to_numeric(lia_prediction)[len(data)]
        pres_lia.append(pred_lia)

    new_list = df1['BS_TOT_LIAB2'].iloc[0:-4].tolist()
    new_list.extend(pres_lia)
    df1['new_lia'] = new_list
    
    data11 = df1['zscore'].iloc[0:-4]
    data22 = df1['zscore'].iloc[0:-3]
    data33 = df1['zscore'].iloc[0:-2]
    data44 = df1['zscore'].iloc[0:-1]
    pres_z = []
    for data in (data11, data22, data33, data44):
        model = AR(data)
        model_fit = model.fit()
        z_prediction = model_fit.predict(len(data), len(data))
        pred_z = pd.to_numeric(z_prediction)[len(data)]
        pres_z.append(pred_z)

    new_list2 = df1['zscore'].iloc[0:-4].tolist()
    new_list2.extend(pres_z)
    df1['new_z'] = new_list2
        
    df_pred = pd.concat([df_pred , df1])
```
