```python
import numpy as np
import pandas as pd
```

```python
D = pd.read_excel ('D.xlsx')
D = D.set_index(pd.to_datetime(D['date']))
D.sort_index(inplace=True)
D[D.company_symbol == 'A'].head(20)
```

```python
S = pd.read_csv ('stock_rate.csv')
S = S.set_index(pd.to_datetime(S['date']))
S.head()
```

```python
data = pd.read_csv('AllData.csv')
data = data.set_index(pd.to_datetime(data['date']))
data.head()
```

```python
####cubic Spline
from scipy.interpolate import interp1d
from scipy import interpolate
```

```python
universe = list(set(D.company_symbol.values))
```

```python
universe.remove('CI')
```

```python
universe[0]
```

```python
D[D.company_symbol == 'AAPL'].head(10)
```

```python
Debt = pd.DataFrame()
for ticker in universe:
    D_test = D[D.company_symbol == ticker]
    upsampled_text = D_test['company_symbol'].resample('D').ffill()
    upsampled_data = D_test[['Current Liabilities','Long-Term Liabilities'
                       ,'Liabilities']].resample('D').interpolate(method='cubic')
# ['company_symbol','Current Liabilities','Long-Term Liabilities','Liabilities']
    merged = pd.merge(upsampled_text, upsampled_data, how='inner', left_index=True, right_index=True)
    Debt = pd.concat([Debt, merged])
```

```python
Debt = Debt.drop(['company_symbol_y'], axis=1)
Debt=data.rename(columns = {'company_symbol_x':'company_symbol'})
```

```python
Debt[Debt.company_symbol == 'AAPL'].head()
```

```python
#combine debt to stock
data = pd.DataFrame()
for ticker in universe:
    S_test = S[S.company_symbol == ticker]
    Debt_test = Debt[Debt.company_symbol == ticker]
    merged =  pd.merge(S_test, Debt_test, how='inner', left_index=True, right_index=True)
    data = pd.concat([data, merged])
data.head()
```

```python
data = data[['change_direction','price','company_symbol_x','r', 
             'Current Liabilities','Long-Term Liabilities','Liabilities']]
data=data.rename(columns = {'company_symbol_x':'company_symbol'})
data.head()
```

```python
returns = pd.DataFrame()
for ticker in universe:
    stotest = data[data.company_symbol == ticker]
    re = pd.Series(np.log(stotest['price'] / stotest['price'] .shift(1,freq = 'B')))
    merged = pd.merge(re, stotest, how='inner', left_index=True, right_index=True)
    returns = pd.concat([returns, merged])
```

```python
returns=returns.rename(columns = {'price_x':'return','price_y':'price'})
```

```python
returns .head()
```

```python
returns[['return']] = returns[['return']].fillna(method='ffill')
```

```python
returns .head(10)
```

```python
sto_vol = pd.DataFrame()
for ticker in universe:
    stotest = returns[returns.company_symbol == ticker]
    vol = stotest['return'].rolling(window=250).std()
    test_merge = pd.merge(stotest, vol, how='inner', left_index=True, right_index=True)
    sto_vol = pd.concat([sto_vol, test_merge])
```

```python
sto_vol= sto_vol.rename(columns = {'return_x':'return','return_y':'vol'})
sto_vol.head()
```

```python
###Merge data with market value of equity###
```

```python
C = pd.read_excel ('capital.xlsx')
C = C.set_index(pd.to_datetime(C['date']))
C.sort_index(inplace=True)
C[C.company_symbol == 'A'].head(5)
```

```python
df = pd.DataFrame()
for ticker in universe:
    voltest = sto_vol[sto_vol.company_symbol == ticker]
    c_test = C[C.company_symbol == ticker]
    test_merge = pd.merge(voltest, c_test, how='inner', left_index=True, right_index=True)
    df = pd.concat([df, test_merge])
```

```python
df.head(300)
```

```python
df = df.drop(['company_symbol_y'], axis=1)
```

```python
df = df.rename(columns = {'company_symbol_x':'company_symbol'})
```

```python
rf = pd.read_excel('risk_free.xlsx')
rf = rf.set_index(pd.to_datetime(rf['date']))
rf.head()
```

```python
rf_up = rf[['risk_free']].resample('D').interpolate(method='cubic')   
```

```python
rf_up.head()
```

```python
alldata = pd.DataFrame()
for ticker in universe:
    test = df[df.company_symbol == ticker]
    test_merge = pd.merge(rf_up, test, how='inner', left_index=True, right_index=True)
    alldata = pd.concat([alldata, test_merge])
```
