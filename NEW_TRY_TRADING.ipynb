```python
def find_abnormal(row):
    mean = np.mean(row)
    if row[-1] > 5*row[0]:
        return 1
    else :
        return 0  
```

```python
for df in (df01, df02, df03, df04, df05):
    universe = list(set(df.Ticker.values))  #########
    lenth = len(universe)
    abn = pd.DataFrame() 
    for ticker in universe:
        df01_ = df[df.Ticker == ticker]
        df01_['abn'] = df01_['dp_stock'].rolling(window='30D',min_periods = 1).apply(find_abnormal)
        abn = pd.concat([abn, df01_])  

    count = 0
    count_indicated_comp = 0
    for ticker in universe:
        df1 = abn[abn.Ticker == ticker] 
        if not df1[(df1.abn == 1) & (df1.date < df01.date[0] + timedelta(days=600))].empty:
            count_indicated_comp +=1
            index = df1[(df1.abn == 1) & (df1.date < df1.date[0] + timedelta(days=600))].index[0]
            df2 = df1[index:]
            if -1 in list(df2['change_direction']):
                count += 1      
        else:
            continue 

    print (count ,lenth, count_indicated_comp  , '\n') 
```
